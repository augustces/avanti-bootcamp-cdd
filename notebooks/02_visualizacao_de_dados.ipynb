# Visualização de Dados

## Importações

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

## Dados para visualização 01

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

## Gráfico que não deve ser feito

```python
ax = sns.barplot(data = df, y = 'arrecadacao', x = 'titulo')
```

## 1. Visualização 01

```python
# usar valores simplificados
df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000)

#ordenar as barras
df = df.sort_values(by = 'arrecadacao', ascending = False)

#usar linhas verticais
sns.set_style('whitegrid')

# usar mesma cor para as barras
# usar barras horizontais
plt.figure(figsize=(5,2))
ax= sns.barplot(data = df, x = "arrecadacao_milhao", y = 'titulo', color = 'tomato')

#adicionar valores nas barras
ax.bar_label(
    ax.containers[0],
    fmt = '%.1f',
    label_type = 'center',
    color = "white"
)

#Melhoria de titulo e labels dos eixos
plt.xlabel('Arrecadacao (Milhões de dólares)')
plt.ylabel('Filme')
plt.title('Arrecadacao de filmes')

#Linha vertical na origem e sem bordas
ax.set(frame_on = False)
ax.axvline(0, color = 'k', clip_on=False)
plt.show()
```

## Dados para visualização 02

```python
df = sns.load_dataset("penguins")
df.head()
```

## Gráfico que não deve ser feito

```python
ax = sns.histplot(data = df, x = 'body_mass_g')
```

## Visualização 02

```python
# usar linhas verticais e horizontais
sns.set_style('whitegrid')

#melhorar titulo e os labels dos eixos
ax = sns.histplot(data = df, x = 'body_mass_g')
plt.ylabel('Quantidade')
plt.xlabel('Massa (g)')
plt.title('Distribuição da massa dos pinguins')

# Linha na origem e remover bordas
ax.set(frame_on = False)
ax.axhline(0, color = 'k', clip_on =False) #linha na origem
plt.show()
```

Dados para visualização 03

```python
df = sns.load_dataset('tips')
df.head()
```

## Boxplot que não deve ser feito

```python
ax = sns.boxplot(x = 'day', y = 'total_bill', data = df)
```

## Visualização 03

```python
# usar linhas verticais
sns.set_style('whitegrid')

# usar mesma cor para as barras
plt.figure(figsize=(6,3))
ax = sns.boxplot(x='day', y ='total_bill', data = df, color = 'cornflowerblue')

# melhorar o titulo e labels dos eixos
plt.xlabel('Dia da semana')
plt.ylabel('Conta (USD)')
plt.title('Distribuição de gorjetas ao longo da semana')

# tirar bordas
ax = ax.set(frame_on=False)

plt.show()
```
